### Data Cleansing

```python
df.drop('LoanID',axis= 1, inplace= True)
df.head()
```

```python
df['EmploymentType'].unique()
```

```python
# convort the categorical into numerical 
from sklearn.preprocessing import LabelEncoder

df['Education'] = LabelEncoder().fit_transform(df['Education'])
df['EmploymentType'] = LabelEncoder().fit_transform(df['EmploymentType'])
df['MaritalStatus'] = LabelEncoder().fit_transform(df['MaritalStatus'])
df['HasMortgage'] = LabelEncoder().fit_transform(df['HasMortgage'])
df['HasDependents'] = LabelEncoder().fit_transform(df['HasDependents'])
df['LoanPurpose'] = LabelEncoder().fit_transform(df['LoanPurpose'])
df['HasCoSigner'] = LabelEncoder().fit_transform(df['HasCoSigner'])

```

```python
df.head(50)
```

### Split the data set into x and y

```python
x = df[['Age', 'Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed',
       'NumCreditLines', 'InterestRate', 'DTIRatio', 'Education',
       'EmploymentType', 'HasMortgage', 'HasDependents',
       'LoanPurpose', 'HasCoSigner']]
print(x.head())
y = df['Default']
y.head()
y.value_counts()
```

### Split the data into training and testing dataset

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size= 0.7, random_state=12350)
```

# Built the Logistic Regression model

```python
def model_LG():
    from sklearn.linear_model import LogisticRegressionCV
 
    model_lg = LogisticRegressionCV().fit(x_train, y_train)
    return model_lg
```

# Built the Naive Bayes 

```python
def model_NB():
    from sklearn.naive_bayes import GaussianNB

    model_nb = GaussianNB().fit(x_train, y_train)
    return model_nb
```

# Built the KNN model

```python
def model_KNN():
    from sklearn.neighbors import KNeighborsClassifier

    model_knn = KNeighborsClassifier().fit(x_train, y_train)
    return model_knn
```

# Built the Support Vector Machine

```python
def model_SVM():
    from sklearn.svm import SVC

    model_svm = SVC(C=3.0).fit(x_train, y_train)
    return model_svm
```

# Built the Decison tree model 

```python
def model_DT():
    from sklearn.tree import DecisionTreeClassifier

    model_dt = DecisionTreeClassifier().fit(x_train, y_train)
    return model_dt
```

# Built the model for Random forest

```python
def model_Rand_forest():
    from sklearn.ensemble import RandomForestClassifier

    model_rand_forest = RandomForestClassifier().fit(x_train, y_train)
    
    return model_rand_forest
```

```python
def model_evaluate(model, model_name):
    from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score,f1_score
    
    
    Accuracy = accuracy_score(y_test, model.predict(x_test))
    Precision = precision_score(y_test, model.predict(x_test),average='macro')
    Recall = recall_score(y_test, model.predict(x_test),average='macro')
    f1_Score = f1_score(y_test, model.predict(x_test),average='macro')

    return model_name,Accuracy,Precision,Recall,f1_Score
```

```python
models = [(model_LG(),"LogisticRegression",(0,0)),(model_KNN(), "KNeighborsClassifier",(0,1)),
          (model_NB(),"Naive Bayes",(1,0)),(model_SVM(),"Support Vector Machine",(1,1)),
          (model_DT(),"DecisionTreeClassifier",(2,0)),(model_Rand_forest(),"RandomForest",(2,1))]
```

# Model Evaluation 

```python
Results = []
for model,model_name,_ in models:
    Results.append(model_evaluate(model, model_name))
```

```python
df1 = pd.DataFrame(Results,columns=['Algorithm','Accuracy', 'Precision', 'Recall', 'F1_Score'])
df1
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, model_Rand_forest().predict(x_test)))
```
